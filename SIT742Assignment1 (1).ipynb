## 2.1 extract high requency words in description

```python
import re
import numpy
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.probability import *
from itertools import chain
#from tqdm import tqdm
import codecs
from sklearn.feature_extraction.text import CountVectorizer
```

```python
with open('stopwords.txt') as f:
    stop_words = f.read().splitlines()
stop_words = set(stop_words)
#stop_words
```

```python
# write your code here
# define your tokenize
import pandas
description  =  data.filter(['description']) #filter the description column from the data table
description




tokenizer = RegexpTokenizer(r"\w+(?:[-']\w+)?")
for x in range(0,len(description)):
 description.iloc[x]['description']  =  tokenizer.tokenize(description.iloc[x]['description']) #tokenizes the description row by row


description['description']  =  description['description'].apply(lambda x: [desc for desc in x if desc not in stop_words])# removes stop words from the description rows 
description



```

```python
# find top common words with document frequencies > 5000
# you may use function FreqDist() and sort()

for x in range(0,len(description)):
 description.iloc[x]['description']=joined_words = ( " ".join(description.iloc[x]['description'])) #joins the tokenized data from description
description

##This code is formed helping from stack overflow a bit of syntax 
from sklearn.feature_extraction.text import CountVectorizer
words  =  CountVectorizer(ngram_range=(1,1), analyzer='word')
matrix  =  words.fit_transform(description.iloc[:,0])
frequencies  =  sum(matrix).toarray()[0]
df  =  pd.DataFrame(frequencies, index=words.get_feature_names(), columns=['frequency'])
pd1  =  df[(df.frequency > 5000)] # stores features with frequency greater than 5000
pd1
    
```

```python
# save your table to 'top_common_words.txt' 
pd1.to_csv('top_common_words.txt', sep='\t', index=True) # saves the dataframe with high frequency words into text file

```

## 2.2 Find key words for describing Shiraz using TF-IDF

```python
# select 'description' from 'variety' eqaul to  'Shiraz'
description1  =  data.filter(['description','variety'])



description1  =  description1[(description1.variety == 'Shiraz')] #filter the description rows with word Shiraz in there description
#wines_new
description_Shiraz  =  description1.filter(['description'])
description_Shiraz
```

```python

from sklearn.feature_extraction.text import TfidfVectorizer

tokenizer = RegexpTokenizer(r"\w+(?:[-']\w+)?")
for x in range(0,len(description_Shiraz)):
 description_Shiraz.iloc[x]['description']  =  tokenizer.tokenize(description_Shiraz.iloc[x]['description'])


description_Shiraz['description']  =  description_Shiraz['description'].apply(lambda x: [desc for desc in x if desc not in stop_words])

for x in range(0,len(description_Shiraz)):
 description_Shiraz.iloc[x]['description']  =  joined_words = ( " ".join(description_Shiraz.iloc[x]['description']))
 

tfidf   =   TfidfVectorizer(analyzer='word', stop_words = 'english')
X   =   tfidf.fit_transform(description_Shiraz.iloc[:,0])

feature   =   tfidf.get_feature_names()


```

```python
# find words with TF-IDF score >0.4 and sort them
X_new = X.multiply(X >= 0.4)  #for filtering tf-idf score >0.4
print(X_new)
y   =   pd.DataFrame(X_new.todense())
y[:1]    =    feature


print(y)
```
